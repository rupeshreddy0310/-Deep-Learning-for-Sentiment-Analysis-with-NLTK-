```python
file="F:\\MS\\SRH study\\III sem\\AI 2\\Novel.txt"
```

```python
# Open the file with explicit encoding
with open(file, 'r', encoding='utf-8') as handle:
    data = handle.read()

# Display the result
print(data)
```

```python
#tokenize the data into features(words);display them
#data pick process, creating a var features
features=word_tokenize(data)  #tokenize splits the words and keeps them seperately
print(features)
```

```python
#normalize the features to lower and exclude punctuation
features=[feature for feature in features if feature.isalpha()]  #
features=[feature.lower() for feature in features]
print(features)
```

```python
#create a list of english stopwords, and then remove them from the featueres
from nltk.corpus import stopwords
stopwords=stopwords.words('english') #which do not have any significance
features=[feature for feature in features if feature not in stopwords]
```

```python
frequencies = FreqDist(features)
plot=frequencies.plot(10)
```

```python
hapaxes=frequencies.hapaxes()
print(hapaxes)
```

```python
ngrams      = ngrams( features,2)
frequencies = FreqDist( ngrams )
frequencies.most_common( 10 )
```

```python
lengths=[len(feature)for feature in features]
print(lengths)
plot=FreqDist(lengths).plot(10)
```

```python
#counting stems=eat(base),eaten as one word  ( run occurs 2 times, running occurs 3 and runned occurs 3.  total run will be 8 times)
from nltk.stem import PorterStemmer #package stem  PorterStemmer is tool
stemmer=PorterStemmer()
stems=[stemmer.stem(feature) for feature in features]  
frequencies=FreqDist(stems)  
frequencies.most_common(10)
```
